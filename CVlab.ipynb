```python
import pandas as pd
# Sample dataset for experiment-1 (rain orediction)
data = [
    ["Sunny", "Warm", "High", "Weak", "Yes"],
    ["Sunny", "Warm", "High", "Strong", "Yes"],
    ["Cloudy", "Warm", "High", "Weak", "Yes"],
    ["Rainy", "Cold", "High", "Weak", "No"],
    ["Sunny", "Warm", "Normal", "Weak", "Yes"],
    ["Cloudy", "Cold", "Normal", "Weak", "No"],
]
columns = ["Sky", "Temperature", "Humidity", "Wind", "Rain"]
df = pd.DataFrame(data, columns = columns)

csv_path = "weather_rain.csv"
df.to_csv(csv_path, index =False)
```
